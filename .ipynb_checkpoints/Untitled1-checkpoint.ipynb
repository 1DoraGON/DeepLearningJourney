```python
import numpy as np  
#from keras.utils import to_categorical  
  
# Load the training set  
with open(r'./data/train-images.idx3-ubyte', 'rb') as f:  
    X_train = np.frombuffer(f.read(), dtype=np.uint8, offset=16).reshape((-1, 28*28))  
with open(r'./data/train-labels.idx1-ubyte', 'rb') as f:  
    y_train = np.frombuffer(f.read(), dtype=np.uint8, offset=8)  
  
# Load the test set  
with open(r'./data/t10k-images.idx3-ubyte', 'rb') as f:  
    X_test = np.frombuffer(f.read(), dtype=np.uint8, offset=16).reshape((-1, 28*28))  
with open(r'./data/t10k-labels.idx1-ubyte', 'rb') as f:  
    y_test = np.frombuffer(f.read(), dtype=np.uint8, offset=8)
```

```python
from keras.datasets import mnist
from keras.utils import to_categorical

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Reshape the data to a 4D tensor - (sample_number, x_img_size, y_img_size, num_channels)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')

# Normalize the pixel values to be between 0 and 1
X_train /= 255
X_test /= 255

# One-hot encode the labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

# Define the model architecture
model = models.Sequential()
model.add(layers.Flatten(input_shape=(28, 28)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
X_train = X_train.reshape((48000, 784))
X_val = X_val.reshape((12000, 784))
X_train.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


# Define the MLP architecture
model = keras.Sequential(
    [
        keras.Input(shape=(784)),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(10, activation="softmax"),
    ]
)

# Compile the model
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using SGD
history_sgd = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=1, verbose=1)

# Compile the model for mini-batch SGD
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)
# Train the model using mini-batch SGD
history_mini_batch_sgd = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=64, verbose=1)

# Compile the model for batch SGD
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using batch SGD
history_batch_sgd = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=len(X_train), verbose=1)

# Plot the learning curves for each strategy (loss and accuracy)
import matplotlib.pyplot as plt

plt.plot(history_sgd.history["loss"], label="SGD")
plt.plot(history_mini_batch_sgd.history["loss"], label="Mini-batch SGD")
plt.plot(history_batch_sgd.history["loss"], label="Batch SGD")
plt.title("Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.plot(history_sgd.history["accuracy"], label="SGD")
plt.plot(history_mini_batch_sgd.history["accuracy"], label="Mini-batch SGD")
plt.plot(history_batch_sgd.history["accuracy"], label="Batch SGD")
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras


# Define the MLP architecture
model = keras.models.Sequential([
    keras.layers.Dense(128, activation="relu", input_shape=(784,)),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

# Compile the model for SGD
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using SGD
history_sgd = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=1, verbose=1)

# Compile the model for mini-batch SGD
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using mini-batch SGD
history_mini_batch_sgd = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=64, verbose=1)

# Compile the model for mini-batch SGD with decay
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01, decay=1e-6),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using mini-batch SGD with decay
history_mini_batch_sgd_decay = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=64, verbose=1)

# Compile the model for SGD with momentum and decay
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, decay=1e-6),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model using SGD with momentum and decay
history_sgd_momentum_decay = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=1, verbose=1)

# Plot the learning curves for all the optimizers
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(history_sgd.history["val_loss"], "r--", label="SGD")
plt.plot(history_mini_batch_sgd.history["val_loss"], "g--", label="Mini-batch SGD")
plt.plot(history_mini_batch_sgd_decay.history["val_loss"], "b--", label="Mini-batch SGD with decay")
plt.plot(history_sgd_momentum_decay.history["val_loss"], "y--", label="SGD with momentum and decay")

plt.title("Validation loss")
plt.legend(loc="best")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

#plt.figure(figsize=(10, 6))

#plt.plot(history_sgd.history["val_accuracy"], "r--)
```
