```python
import numpy as np
from keras.datasets import mnist
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess the data by normalizing the pixel values to be between 0 and 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape the data to be compatible with the Keras API
X_train = X_train.reshape((X_train.shape[0], 28, 28, 1))
X_test = X_test.reshape((X_test.shape[0], 28, 28, 1))

# Convert the class labels to one-hot encoded vectors
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Split the training set into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.optimizers import Adam


# Define the model architecture
model1 = Sequential()
model1.add(Flatten(input_shape=(28, 28, 1)))
model1.add(Dense(128, activation='relu'))
model1.add(Dense(64, activation='relu'))
model1.add(Dense(10, activation='softmax'))


model2 = Sequential()
model2.add(Flatten(input_shape=(28, 28, 1)))
model2.add(Dense(128, activation='relu'))
model2.add(Dense(64, activation='relu'))
model2.add(Dense(10, activation='softmax'))


model3 = Sequential()
model3.add(Flatten(input_shape=(28, 28, 1)))
model3.add(Dense(128, activation='relu'))
model3.add(Dense(64, activation='relu'))
model3.add(Dense(10, activation='softmax'))

# Compile the model
model1.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
import matplotlib.pyplot as plt
epochs = 10
# Train the model using SGD
batch_sgd_history = model1.fit(X_train, y_train, batch_size=len(X_train), epochs=epochs, validation_data=(X_val, y_val), verbose=1)
model2.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model using mini-batch SGD
mini_batch_sgd_history = model2.fit(X_train, y_train, batch_size=64, epochs=epochs, validation_data=(X_val, y_val), verbose=1)
model3.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model using batch SGD
sgd_history = model3.fit(X_train, y_train, batch_size=1, epochs=epochs, validation_data=(X_val, y_val), verbose=1)

# Plot the learning curves for each strategy
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(sgd_history.history['loss'], label='SGD')
plt.plot(mini_batch_sgd_history.history['loss'], label='Mini-batch SGD')
plt.plot(batch_sgd_history.history['loss'], label='Batch SGD')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(sgd_history.history['accuracy'], label='SGD')
plt.plot(mini_batch_sgd_history.history['accuracy'], label='Mini-batch SGD')
plt.plot(batch_sgd_history.history['accuracy'], label='Batch SGD')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
from keras.optimizers import SGD
from keras.callbacks import LearningRateScheduler

# Define learning rate decay function
def lr_decay(epoch):
    initial_learning_rate = 0.01
    decay_rate = 1e-6
    return initial_learning_rate / (1 + decay_rate * epoch)

# Create model
model_mini_batch_sgd_decay = Sequential()
model_mini_batch_sgd_decay.add(Flatten(input_shape=(28, 28, 1)))
model_mini_batch_sgd_decay.add(Dense(128, activation='relu'))
model_mini_batch_sgd_decay.add(Dense(64, activation='relu'))
model_mini_batch_sgd_decay.add(Dense(10, activation='softmax'))

# Compile the model
sgd_with_decay = SGD(lr=0.01, decay=1e-6)
model_mini_batch_sgd_decay.compile(optimizer=sgd_with_decay, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model using mini-batch SGD with decay
batch_size = 64
epochs = 10

lr_scheduler = LearningRateScheduler(lr_decay)

mini_batch_sgd_decay_history = model_mini_batch_sgd_decay.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_val, y_val), callbacks=[lr_scheduler], verbose=1)

```

```python
# Create model
model_sgd_decay_momentum = Sequential()
model_sgd_decay_momentum.add(Flatten(input_shape=(28, 28, 1)))
model_sgd_decay_momentum.add(Dense(128, activation='relu'))
model_sgd_decay_momentum.add(Dense(64, activation='relu'))
model_sgd_decay_momentum.add(Dense(10, activation='softmax'))

# Compile the model
sgd_with_decay_momentum = SGD(lr=0.01, decay=1e-6, momentum=0.9)
model_sgd_decay_momentum.compile(optimizer=sgd_with_decay_momentum, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model using SGD with decay and momentum
batch_size = len(X_train)
epochs = 10

sgd_decay_momentum_history = model_sgd_decay_momentum.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_val, y_val), verbose=1)

```
